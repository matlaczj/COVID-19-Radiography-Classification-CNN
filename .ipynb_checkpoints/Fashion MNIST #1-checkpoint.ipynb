```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
print(tf.__version__)
```

**Fashion MNIST, który zawiera 70 000 obrazów w skali szarości w 10 kategoriach. Zdjęcia przedstawiają poszczególne elementy odzieży w niskiej rozdzielczości (28 na 28 pikseli). 60 000 obrazów jest używanych do uczenia sieci i 10 000 obrazów do oceny.**

```python
fashion_mnist = tf.keras.datasets.fashion_mnist
(train_images, train_labels), (test_images, test_labels) = fashion_mnist.load_data()
```

```python
print("train_images")
print("ndim",train_images.ndim)
print("shape",train_images.shape)
print("\ntrain_labels")
print("ndim",train_labels.ndim)
print("shape",train_labels.shape)
print("\ntest_images")
print("ndim",test_images.ndim)
print("shape",test_images.shape)
print("\ntest_labels")
print("ndim",test_labels.ndim)
print("shape",test_labels.shape)
```

**Obrazy mają wymiary 28x28 NumPy, z wartościami pikseli w zakresie od 0 do 255. Etykiety są tablicą liczb całkowitych z zakresu od 0 do 9. Odpowiadają one klasie odzieży, którą przedstawia obraz.** <br>
0	T-shirt / top <br>
1	Spodnie <br>
2	Sweter <br>
3	Sukienka <br>
4	Płaszcz <br>
5	Sandał <br>
6	Koszula <br>
7	Sneaker <br>
8	Torba <br>
9	But do kostki <br>

```python
class_names = ['T-shirt/top', 'Trouser', 'Pullover', 'Dress', 'Coat', 'Sandal', 'Shirt', 'Sneaker', 'Bag', 'Ankle boot']
```

**Oglądanie zbioru**

```python
def show_data(data,labels,class_names,size,nrows,ncols):
    plt.figure(figsize=(10,10))
    for i in range(nrows*ncols):
        plt.subplot(nrows,ncols,i+1)
        plt.xticks([])
        plt.yticks([])
        plt.grid(False)
        plt.imshow(data[i], cmap=plt.cm.binary)
        plt.xlabel(class_names[labels[i]])
    plt.show()
```

```python
show_data(train_images,train_labels,class_names,size=10,nrows=6,ncols=6)
```

**Skalowanie** do zakresu <0,1>

```python
train_images = train_images / 255.0
test_images = test_images / 255.0
```

**Budowa modelu**, a następnie jego kompilacja.

```python
model = tf.keras.Sequential([
    tf.keras.layers.Flatten(input_shape=(28, 28)),
    tf.keras.layers.Dense(128, activation='relu'),
    tf.keras.layers.Dropout(0.2),
    tf.keras.layers.Dense(10)
])
```

```python
model.compile(optimizer='SGD',
              loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
              metrics=['accuracy'])
```

**Callback**, pod koniec każdej epoki zmniejsza *learning_rate*.

```python
class LearningRateReducerCb(tf.keras.callbacks.Callback):
    def on_epoch_end(self, epoch, logs={}):
        skalar = 0.99
        old_lr = self.model.optimizer.lr.read_value()
        new_lr = old_lr * skalar
        print("\n>> Reducing Learning Rate from {} to {}\n".format(old_lr, new_lr))
        self.model.optimizer.lr.assign(new_lr)
```

```python
class LossHistoryCb(tf.keras.callbacks.Callback):
    def on_train_begin(self, logs={}):
        self.losses = []
        self.accuracies = []
    def on_batch_end(self, batch, logs={}):
        self.losses.append(logs.get('loss'))
        self.accuracies.append(logs.get('accuracy'))
high_res_history = LossHistoryCb();
```

**Trenowanie**, czyli dopasowywanie modelu do danych.

```python
history = model.fit(train_images, 
                    train_labels, 
                    epochs=10, 
                    batch_size=128, 
                    callbacks=[LearningRateReducerCb(), high_res_history], 
                    workers=2,
                    use_multiprocessing=True,
                    max_queue_size=100,
                   validation_split=0.1)
```

**Wizualizacja historii trenowania**

```python
print(history.history.keys())
```

```python
loss = history.history['loss']
accuracy = history.history['accuracy']
val_loss = history.history['val_loss']
val_accuracy = history.history['val_accuracy']

plt.figure(figsize=(15,5))
plt.subplot(1,2,1)
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.title('Training and validation Loss')
plt.plot(loss,'bo:',label="Traning Loss")
plt.plot(val_loss,'go:',label="Validation Loss")
plt.legend()

plt.subplot(1,2,2)
plt.xlabel("Epoch")
plt.ylabel("Accuracy")
plt.title('Training and validation Accuracy')
plt.plot(accuracy,'bo:',label="Training Accuracy")
plt.plot(val_accuracy,'go:',label="Validation Accuracy")
plt.legend()
```

**Wykresy co batch**, zamiast tylko co epokę, co daje znacznie większą rozdzielczość. <br>
Chwilowe wyniki typu 96% mogą zmylać, bo jest to wynik w pojedynczym batchu, a nie w całej epoce (zbiorze).

```python
plt.figure(figsize=(15,5))
plt.subplot(1,2,1)
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.title('Training Loss')
plt.plot(high_res_history.losses,'b',label="Traning Loss")
plt.legend()

plt.subplot(1,2,2)
plt.xlabel("Epoch")
plt.ylabel("Accuracy")
plt.title('Training Accuracy')
plt.plot(high_res_history.accuracies,'b',label="Traning Accuracy")
plt.legend()
```

**Testowanie**, czyli ocena dokładności.

```python
test_loss, test_acc = model.evaluate(test_images,  test_labels, verbose=2)
print('\nTest accuracy:', test_acc)
```

**Korzystanie z modelu**, do przewidywania dla nowych danych.

```python
probability_model = tf.keras.Sequential([model,tf.keras.layers.Softmax()])
```

```python
predictions = probability_model.predict(test_images)
```

**Oglądanie prognoz modelu**

```python
def plot_image(i, predictions_array, true_label, img):
  true_label, img = true_label[i], img[i]
  plt.grid(False)
  plt.xticks([])
  plt.yticks([])

  plt.imshow(img, cmap=plt.cm.binary)

  predicted_label = np.argmax(predictions_array)
  if predicted_label == true_label:
    color = 'blue'
  else:
    color = 'red'

  plt.xlabel("{} {:2.0f}% ({})".format(class_names[predicted_label],
                                100*np.max(predictions_array),
                                class_names[true_label]),
                                color=color)

def plot_value_array(i, predictions_array, true_label):
  true_label = true_label[i]
  plt.grid(False)
  plt.xticks(range(10))
  plt.yticks([])
  thisplot = plt.bar(range(10), predictions_array, color="#777777")
  plt.ylim([0, 1])
  predicted_label = np.argmax(predictions_array)

  thisplot[predicted_label].set_color('red')
  thisplot[true_label].set_color('blue')
```

```python
num_rows = 5
num_cols = 3
num_images = num_rows*num_cols
plt.figure(figsize=(2*2*num_cols, 2*num_rows))
for i in range(num_images):
  plt.subplot(num_rows, 2*num_cols, 2*i+1)
  plot_image(i, predictions[i], test_labels, test_images)
  plt.subplot(num_rows, 2*num_cols, 2*i+2)
  plot_value_array(i, predictions[i], test_labels)
plt.tight_layout()
plt.show()
```
